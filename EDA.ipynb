For simplifying the dataset:
* Gearbox: Automata (x fokozat) --> Automata, Manualis (x fokozat) --> Manualis
* Aircon_type: Digital * --> Digital,

```python
cars_df_cleaned['Gearbox_short'] = cars_df_cleaned['Gearbox'].apply(lambda x: 'Automata' if 'automata' in str(x).lower() else x)
cars_df_cleaned['Gearbox_short'] = cars_df_cleaned.apply(lambda x: 'Szekvenciális' if 'szekvenciális' in str(x.Gearbox).lower() else x.Gearbox_short, axis = 1)
cars_df_cleaned['Gearbox_short'] = cars_df_cleaned.apply(lambda x: 'Manuális' if 'manuális' in str(x.Gearbox).lower() else x.Gearbox_short, axis = 1)

cars_df_cleaned['AirCon_Type_short'] = cars_df_cleaned['AirCon_Type'].apply(lambda x: 'Digitális' if 'digitális' in str(x).lower() else x)
cars_df_cleaned['AirCon_Type_short'] = cars_df_cleaned.apply(lambda x: 'Automata' if 'automata' in str(x.AirCon_Type).lower() else x.AirCon_Type_short, axis = 1)
cars_df_cleaned['AirCon_Type_short'] = cars_df_cleaned.apply(lambda x: 'Manuális' if 'manuális' in str(x.AirCon_Type).lower() else x.AirCon_Type_short, axis = 1)
```
