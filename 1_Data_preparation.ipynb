### 2. Data cleaning

Check if we have duplicate rows:

```python
cars_df[cars_df.duplicated(keep=False)]
```

```python
# Drop duplicated
cars_df = cars_df.drop_duplicates()
```

```python
# Double check
cars_df[cars_df.duplicated(keep=False)]
```

We can drop the columns where there are several missing values, or those which are just irrelevant for us, for instance EUR price. (we have HUF price already)

```python
irrelevant_cols = ['Vételár EUR:', 'Egyéb költségek:','Fizetendő magyarországi forgalomba helyezés esetén:'\
                   ,'Járműelőélet:','Kárpit színe (1):','Kárpit színe (2):','Kezdőrészlet:','Havi részlet:'\
                   ,'Finanszírozás:','Nyári gumi méret:','Téli gumi méret:','Futamidő:','Tető:'\
                   ,'Havi részlet:','Hátsó nyári gumi méret:','Hátsó téli gumi méret:','Bérlési lehetőség:']

cars_df_cleaned = cars_df.drop(axis = 1, columns=irrelevant_cols)
```

```python
cars_df_cleaned.info()
```

We are going to deal with the missing values in the feature engineering part. 

3 cars had discount price (Akciós ár:), I replaced the original price with these values. After that, I deleted the discount price columns.

```python
cars_df_cleaned['final_price'] = cars_df_cleaned[['Vételár:','Akciós ár:']].\
                                 apply(lambda x: x[0] if pd.isnull(x[1]) else x[1], axis=1)
```

##### 'final_price' column

'2 999 000 Ft' ➡ '2999000'

```python
# final_price encoding issue fix
cars_df_cleaned.loc[0,'final_price']
```

```python
# \xa0 --> to ''
cars_df_cleaned['final_price'] = cars_df_cleaned['final_price'].apply(lambda x: x.replace(u'\xa0', u''))
```

```python
# Exlude 'Ft'
cars_df_cleaned['final_price'] = cars_df_cleaned['final_price'].apply(lambda x: x[:-2])

# Parse to numeric
cars_df_cleaned['final_price'] = pd.to_numeric(cars_df_cleaned['final_price'])
```

```python
# Drop Vételár: and Akciós ár: columns
cars_df_cleaned.drop(columns=['Vételár:','Akciós ár:'], inplace=True)
```

#### 'Évjárat:' column

'2012/8	' ➡ '2012'

'1990 30 évnél öregebb' ➡ '1990'

```python
# Display only the year
cars_df_cleaned['Évjárat:'] = cars_df_cleaned['Évjárat:'].apply(lambda x: x.split('/')[0].strip())

cars_df_cleaned['Évjárat:'] = cars_df_cleaned['Évjárat:'].apply(lambda x: x.split(' ')[0].strip() if ' ' in x else x)
```

#### 'Km. óra állás:' column

'245 000 km' ➡ '245000'

```python
# Fix encoding issue
cars_df_cleaned['Km. óra állás:'] = cars_df_cleaned['Km. óra állás:'].apply(lambda x: x.replace(u'\xa0', u'')[:-2])

# Parse to numeric
cars_df_cleaned['Km. óra állás:'] = pd.to_numeric(cars_df_cleaned['Km. óra állás:'])
```

#### 'Száll. szem szám:' + 'Ajtók száma:' column

'5 fő' ➡ '5'

```python
# Exclude 'fő'
cars_df_cleaned['Szállítható szem. száma:'] = cars_df_cleaned['Szállítható szem. száma:']\
                                              .apply(lambda x: str(x)[:-2].strip() if not pd.isnull(x) else x)

# Parse to numeric 1
#cars_df_cleaned['Szállítható szem. száma:'] = pd.to_numeric(cars_df_cleaned['Szállítható szem. száma:'])

# Parse to numeric 2
#cars_df_cleaned['Ajtók száma:'] = pd.to_numeric(cars_df_cleaned['Ajtók száma:'])
```

#### 'Tömeg:' columns

'1 338 kg' ➡ '1338'

```python
# Fix encoding issue
cars_df_cleaned['Saját tömeg:'] = cars_df_cleaned['Saját tömeg:']\
                                  .apply(lambda x: str(x).replace(u'\xa0', u'')[:-2] if x==x  else x)

cars_df_cleaned['Teljes tömeg:'] = cars_df_cleaned['Teljes tömeg:']\
                                  .apply(lambda x: str(x).replace(u'\xa0', u'')[:-2] if x==x  else x)

# Parse to numeric
cars_df_cleaned['Saját tömeg:'] = pd.to_numeric(cars_df_cleaned['Saját tömeg:'])

cars_df_cleaned['Teljes tömeg:'] = pd.to_numeric(cars_df_cleaned['Teljes tömeg:'])
```

#### 'Csomagtartó:' column

'316 liter' ➡ '316'

```python
# Exclude 'liter'
cars_df_cleaned['Csomagtartó:'] = cars_df_cleaned['Csomagtartó:'].apply(lambda x: str(x)[:-5] if x==x  else x)

cars_df_cleaned['Csomagtartó:'] = cars_df_cleaned['Csomagtartó:']\
                                  .apply(lambda x: str(x).replace(u'\xa0', u'') if x==x  else x)

# Parse to numeric
cars_df_cleaned['Csomagtartó:'] = pd.to_numeric(cars_df_cleaned['Csomagtartó:'])
```

#### 'Hengerűrtartalom:' column

'1 560 cm³' ➡ '1560'

```python
# Fix encoding issue
cars_df_cleaned['Hengerűrtartalom:'] = cars_df_cleaned['Hengerűrtartalom:']\
                                       .apply(lambda x: x.replace(u'\xa0', u'').split(' ')[0] if x==x else x)

# Parse to numeric
cars_df_cleaned['Hengerűrtartalom:'] = pd.to_numeric(cars_df_cleaned['Hengerűrtartalom:'])
```

#### 'Teljesítmény:' column

This column should be splitted into 2 new columns, Power and Horsepower.

'70 kW, 95 LE' ➡ '70 kW','95 LE' ➡ '70',95'

```python
# New columns
cars_df_cleaned['Teljesítmény (kW)'] = cars_df_cleaned['Teljesítmény:'].apply(lambda x: str(x).split(',')[0] if x==x else x)
cars_df_cleaned['Teljesítmény (LE)'] = cars_df_cleaned['Teljesítmény:'].apply(lambda x: str(x).split(',')[1] if x==x else x)

# Exclude kW and LE
cars_df_cleaned['Teljesítmény (kW)'] = cars_df_cleaned['Teljesítmény (kW)'].apply(lambda x: x.replace(u'\xa0', u'') if x == x else x)
cars_df_cleaned['Teljesítmény (LE)'] = cars_df_cleaned['Teljesítmény (LE)'].apply(lambda x: x.replace(u'\xa0', u'') if x==x else x)

cars_df_cleaned['Teljesítmény (kW)'] = cars_df_cleaned['Teljesítmény (kW)'].apply(lambda x: x[:-2] if x==x else x)
cars_df_cleaned['Teljesítmény (LE)'] = cars_df_cleaned['Teljesítmény (LE)'].apply(lambda x: x[:-2] if x==x else x)

# Parse to numeric
cars_df_cleaned['Teljesítmény (kW)'] = pd.to_numeric(cars_df_cleaned['Teljesítmény (kW)'])
cars_df_cleaned['Teljesítmény (LE)'] = pd.to_numeric(cars_df_cleaned['Teljesítmény (LE)'])

# Drop 'Teljesítmény:' column
cars_df_cleaned.drop(columns='Teljesítmény:', axis=1, inplace=True)
```

#### 'Műszaki vizsga érvényes:' column

'2024/8' ➡ '2024'

```python
cars_df_cleaned['Műszaki vizsga érvényes:'] = cars_df_cleaned['Műszaki vizsga érvényes:']\
                                              .apply(lambda x: str(x).split('/')[0])
```

#### Marka column

I wanted to show only the manufacturer name, for simplicity.

'FORD FOCUS 1.6 TDCi Trend Techno érvényes vizsga' --> 'FORD'

```python
# Get only the manufacturer name
cars_df_cleaned['Marka'] = cars_df_cleaned['Marka'].apply(lambda x: x.split(' ')[0])
```

#### Last but not least reorder the columns and change the column names

```python
cols = cars_df_cleaned.columns.tolist()

cols = [cols[-3]] + cols[:-3] + cols[-2:]

cars_df_cleaned = cars_df_cleaned[cols]
```

```python
colname_dict = {
    'final_price':'Price (HUF)',
    'Marka':'Brand',
    'Évjárat:':'Year',
    'Állapot:':'Condition',
    'Kivitel:':'Design',
    'Km. óra állás:':'Kilometers',
    'Szállítható szem. száma:':'Persons',
    'Ajtók száma:':'Doors',
    'Szín:':'Colour',
    'Saját tömeg:':'Own_weight (kg)',
    'Teljes tömeg:':'Full_weight (kg)',
    'Csomagtartó:':'Trunk (liter)',
    'Klíma fajtája:':'AirCon_Type',
    'Üzemanyag:':'Fuel',
    'Hengerűrtartalom:':'Cylinder (cm3)',
    'Henger-elrendezés:':'Layout',
    'Hajtás:':'Wheel_drive',
    'Sebességváltó:':'Gearbox',
    'Okmányok jellege:':'Document',
    'Műszaki vizsga érvényes:':'Roadworthiness',
    'Teljesítmény (kW)':'Power (kW)',
    'Teljesítmény (LE)':'HorsePower (hp)'
}

cars_df_cleaned.rename(columns=colname_dict, inplace=True)
```

#### Create age column for the further analysis

```python
cars_df_cleaned['Age'] = datetime.today().year - pd.to_numeric(cars_df_cleaned['Year'])
```
