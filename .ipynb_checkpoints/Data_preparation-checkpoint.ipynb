```python
import pandas as pd
import numpy as np
```

```python
cars_df = pd.read_pickle('./dataset/cars_df.pkl')
```

```python
# Raw data
cars_df.head()
```

```python
cars_df.info()
```

```python
# DROP vetelar EUR, finanszirozas, sajat tomeg, fizetendő, egyéb, bérlési leh,  X
# vételár --> nan akciós ár
# marka csak tipus
# vétel ár to int
# évjárat csak /-ig
# km to int
# szállíthato szemely es ajto to int
```

```python
# drop columns where there are several missing values, or those which are just irrelevant for us, for instance EUR price (we have HUF price already)
irrelevant_cols = ['Vételár EUR:','Finanszírozás:','Kárpit színe (1):','Műszaki vizsga érvényes:','Nyári gumi méret:','Alvázszám:','Kárpit színe (2):','Tető:','Téli gumi méret:','Hátsó nyári gumi méret:','Hátsó téli gumi méret:','Kezdőrészlet:','Futamidő:','Bérlési lehetőség:','Egyéb költségek:','Fizetendő magyarországi forgalomba helyezés esetén:']
cars_df_cleaned = cars_df.drop(axis = 1, columns=irrelevant_cols)
```

```python
cars_df_cleaned.info()
```

```python
# 18 cars where there was a discount price --> replace the original price with the discounted one

cars_df_cleaned.loc[~cars_df_cleaned['Akciós ár:'].isnull(), 'Vételár:'] = cars_df_cleaned.loc[~cars_df_cleaned['Akciós ár:'].isnull(), 'Akciós ár:']

# we can detele discount price

cars_df_cleaned.drop(columns=['Akciós ár:'], axis = 1, inplace=True)
```

```python
cars_df_cleaned.head()
```

### Vételár: column

```python
# Vételár coding issue should be fixed
cars_df_cleaned.iloc[0,1]
# \xa0 --> to ''
cars_df_cleaned['Vételár:'] = cars_df_cleaned['Vételár:'].apply(lambda x: x.replace(u'\xa0', u''))
```

```python
# we dont need Ft
cars_df_cleaned['Vételár:'] = cars_df_cleaned['Vételár:'].apply(lambda x: x[:-2])

# to numeric
cars_df_cleaned['Vételár:'] = pd.to_numeric(cars_df_cleaned['Vételár:'])
```

### Évjárat: column

```python
cars_df_cleaned['Évjárat:'] = cars_df_cleaned['Évjárat:'].apply(lambda x: x.split('/')[0].strip())
```

### KM column

```python
cars_df_cleaned['Kilométeróra állása:'] = cars_df_cleaned['Kilométeróra állása:'].apply(lambda x: x.replace(u'\xa0', u'')[:-2])
cars_df_cleaned['Kilométeróra állása:'] = pd.to_numeric(cars_df_cleaned['Kilométeróra állása:'])
```

### 'Száll. szem szám:' + 'Ajtók száma:' column

```python
# exclude 'fő', if x is not NaN (x!=x)
cars_df_cleaned['Szállítható szem. száma:'] = cars_df_cleaned['Szállítható szem. száma:'].apply(lambda x: str(x)[:-2].strip() if x==x else x)

# to numeric
cars_df_cleaned['Szállítható szem. száma:'] = pd.to_numeric(cars_df_cleaned['Szállítható szem. száma:'])

# Ajtók száma: to numeric
cars_df_cleaned['Ajtók száma:'] = pd.to_numeric(cars_df_cleaned['Ajtók száma:'])
```

### Tömeg: columns

```python
cars_df_cleaned['Saját tömeg:'] = cars_df_cleaned['Saját tömeg:'].apply(lambda x: str(x).replace(u'\xa0', u'')[:-2] if x==x  else x)
cars_df_cleaned['Teljes tömeg:'] = cars_df_cleaned['Teljes tömeg:'].apply(lambda x: str(x).replace(u'\xa0', u'')[:-2] if x==x  else x)

cars_df_cleaned['Saját tömeg:'] = pd.to_numeric(cars_df_cleaned['Saját tömeg:'])
cars_df_cleaned['Teljes tömeg:'] = pd.to_numeric(cars_df_cleaned['Teljes tömeg:'])
```

### Csomagtartó: column

```python
cars_df_cleaned['Csomagtartó:'] = cars_df_cleaned['Csomagtartó:'].apply(lambda x: str(x)[:-5] if x==x  else x)

cars_df_cleaned['Csomagtartó:'] = pd.to_numeric(cars_df_cleaned['Csomagtartó:'])
```

### Hengerűrtartalom column

```python
cars_df_cleaned['Hengerűrtartalom:'] = cars_df_cleaned['Hengerűrtartalom:'].apply(lambda x: x.replace(u'\xa0', u'').split(' ')[0] if x==x else x)

cars_df_cleaned['Hengerűrtartalom:'] = pd.to_numeric(cars_df_cleaned['Hengerűrtartalom:'])
```

### Teljesítmény split into 2 columns

```python
cars_df_cleaned['Teljesítmény (kW)'] = cars_df_cleaned['Teljesítmény:'].apply(lambda x: x.split(',')[0])
cars_df_cleaned['Teljesítmény (LE)'] = cars_df_cleaned['Teljesítmény:'].apply(lambda x: x.split(',')[1])

cars_df_cleaned['Teljesítmény (kW)'] = cars_df_cleaned['Teljesítmény (kW)'].apply(lambda x: x.split(' ')[0])
cars_df_cleaned['Teljesítmény (LE)'] = cars_df_cleaned['Teljesítmény (LE)'].apply(lambda x: x.split(' ')[1])

cars_df_cleaned['Teljesítmény (kW)'] = pd.to_numeric(cars_df_cleaned['Teljesítmény (kW)'])
cars_df_cleaned['Teljesítmény (LE)'] = pd.to_numeric(cars_df_cleaned['Teljesítmény (LE)'])

cars_df_cleaned.drop(columns='Teljesítmény:', axis=1, inplace=True)
```

### Show only the brand of the car for simplicity

```python
cars_df_cleaned['Márka'] = cars_df_cleaned['Marka'].apply(lambda x: x.split(' ')[0])
```

### Finally reorder the columns and change the column names

```python
cols = cars_df_cleaned.columns.tolist()

cols = [cols[0]] + cols[-1:] + cols[1:-1]

cars_df_cleaned = cars_df_cleaned[cols]
```

```python
colname_dict = {
    'Marka':'Full_Name',
    'Márka':'Brand',
    'Vételár:':'Price (HUF)',
    'Évjárat:':'Year',
    'Állapot:':'Condition',
    'Kivitel:':'Design',
    'Kilométeróra állása:':'Kilometers',
    'Szállítható szem. száma:':'Persons',
    'Ajtók száma:':'Doors',
    'Szín:':'Colour',
    'Saját tömeg:':'Own_weight (kg)',
    'Teljes tömeg:':'Full_weight (kg)',
    'Csomagtartó:':'Trunk (liter)',
    'Klíma fajtája:':'AirCon_Type',
    'Üzemanyag:':'Fuel',
    'Hengerűrtartalom:':'Cylinder (cm3)',
    'Henger-elrendezés:':'Layout',
    'Hajtás:':'Wheel_drive',
    'Sebességváltó fajtája:':'Gearbox',
    'Okmányok jellege:':'Document',
    'Teljesítmény (kW)':'Power (kW)',
    'Teljesítmény (LE)':'HorsePower (hp)'
}

cars_df_cleaned.rename(columns=colname_dict, inplace=True)
```
